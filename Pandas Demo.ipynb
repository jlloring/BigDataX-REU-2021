```python
d_parser = lambda x: pd.datetime.strptime(x, '%Y-%m-%d %I-%p') #creates function
df_ts = pd.read_csv('ETH_1h.csv', parse_dates=['Date'], date_parser = d_parser) #imports w function
df_ts.head()
```

```python
df_ts.shape #gives us more info on the df_ts
```

```python
df_ts.loc[0,'Date'].day_name() #need to convert it to a date-time (if not importing with function)
```

```python
df_ts['Date'] = pd.to_datetime(df_ts['Date']) #pandas doesn't recognize the date format
                                                #of the df_ts
```

```python
df_ts['Date'] = pd.to_datetime(df_ts['Date'], format = '%Y-%m-%d %I-%p') #correctly converts
                                                                             #to date-time
```

```python
df_ts['Date']
```

```python
df_ts.loc[0,'Date'].day_name() #tells us the day of the week of the first date
                                #now that it's converted to the correct format
```

```python
df_ts['Date'].dt.day_name() #tells us the day of the weeks for each date value in the series
```

```python
df_ts['DayOfWeek'] = df_ts['Date'].dt.day_name() #creates new DayOfWeek column
df_ts
```
