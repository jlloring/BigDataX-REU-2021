```python
df_1 = pd.read_csv('data1.csv') #imports csv file
```

```python
ids = df_1['Responder_id'] #creates ids object of all responder ids
lang_responses = df_1['LanguagesWorkedWith'] #creates object of all languages worked with
```

```python
from collections import Counter #did stuff on Mac terminal first
language_counter = Counter() #creating an object of an empty counter
```

```python
for response in lang_responses:
    language_counter.update(response.split(';'))
```

```python
languages = [] #creating empty object
popularity = [] #creating empty object

for item in language_counter.most_common(15):
    languages.append(item[0])
    popularity.append(item[1])
```

```python
# Creating horizontal bar chart (depends on 5 above cells)
plt.barh(languages,popularity) #creates horizontal bar chart

plt.title("Most Popular Languages")
plt.xlabel("Number of People Who Use")

plt.tight_layout()
plt.grid(True)
plt.show()
```

```python
# Creating a pie chart
plt.style.use('fivethirtyeight') #setting the style for our pie chart

slices = [59219, 55466, 47544, 36443, 35917] #slices for making the pie chart
labels = ['JavaScript', 'HTML/CSS', 'SQL', 'Python', 'Java']

explode =[0,0,0,0.1,0] #creates a list for explosion

plt.pie(slices, labels=labels, explode=explode, shadow=True, 
        startangle=90, autopct='%1.1f%%', wedgeprops={'edgecolor':'black'})
    #makes the pie chart of the slices object, uses some different arguments

plt.title("My Awesome Pie Chart")
plt.tight_layout()
plt.show()
```

```python
# Creating a stack plot
plt.style.use('fast') #setting the style for our stack plot

minutes = [1,2,3,4,5,6,7,8,9] #provided data

player1 = [8,6,5,5,4,2,1,1,0] #provided data
player2 = [0,1,2,2,2,4,4,4,4] #provided data
player3 = [0,1,1,1,2,2,3,3,4] #provided data

labels=['player1','player2','player3'] #creates labels list
colors = ['#6d904f','#fc4f30','#008fd5'] #creates colors list

plt.stackplot(minutes, player1, player2, player3, labels=labels, colors=colors)
    #creates stack plot with some arguments

#plt.legend(loc='lower left')
plt.legend(loc=(0.07,0.05)) #uses coordinates for the legend

plt.title("My Awesome Stack Plot")
plt.tight_layout()
plt.show()
```

```python
df_2 = pd.read_csv('data2.csv') #imports data
df_2.head() #prints first 5 rows
```

```python
ages = df_2['Age'] #creating Age variable from the Age column

dev_salaries = df_2['All_Devs'] #creating dev_salaries variable from that column

py_salaries = df_2['Python'] #creating py_salaries variable from that column

js_salaries = df_2['JavaScript'] #creating js_salaries variable from that column
```

```python
# Filling Area on Line Plots

plt.plot(ages, dev_salaries, color='#444444', linestyle='--',
         label='All Devs') #creates first plot

plt.plot(ages, py_salaries, label='Python') #creates second plot

overall_median = 57287

plt.fill_between(ages, py_salaries, dev_salaries, 
                 where=(py_salaries > dev_salaries), 
                 interpolate=True, alpha=0.25, label='Above Avg') 
    # the argument after py_salaries (before next""=) is set to 0 by default,
        #unless otherwise specified (here it's dev_salaries)
        
plt.fill_between(ages, py_salaries, dev_salaries, 
                 where=(py_salaries <= dev_salaries), 
                 interpolate=True, color='red', alpha=0.25, label='Below Avg')
    # creates another fill_between!

plt.legend()

plt.title('Median Salary (USD) by Age')
plt.xlabel('Ages')
plt.ylabel('Median Salary (USD)')

plt.show()
```

```python
df_6 = pd.read_csv('data6.csv') #imports data
df_6.head() #prints first 5 rows
```

```python
ages = df_6['Age']
dev_salaries = df_6['All_Devs']
py_salaries = df_6['Python']
js_salaries = df_6['JavaScript']
```

```python
# SubPlotting (imported data, 2 plots on the same image)
plt.style.use('seaborn') #setting plot style

fig, (ax1, ax2) = plt.subplots(nrows=2, ncols=1, sharex=True) #setting up subplotting, changed plt.plot to ax.plot
    #this variable is simply equal to a subplot

ax1.plot(ages, dev_salaries, color='#444444', linestyle='--', label='All Devs')
    
ax2.plot(ages, py_salaries, label='Python')
ax2.plot(ages, js_salaries, label='JavaScript')

ax1.legend()
ax1.set_title('Median Salary (USD) by Age') #changed to ax and added a set_
ax1.set_ylabel('Median Salary (USD)') #changed to ax and added a set_

ax2.legend()
ax2.set_xlabel('Ages')
ax2.set_ylabel('Median Salary (USD)')

plt.tight_layout()
plt.show()
```

```python
# SubPlotting (imported data, 2 plots on separate figures)
plt.style.use('seaborn') #setting plot style

fig1, ax1 = plt.subplots()
fig2, ax2 = plt.subplots()

ax1.plot(ages, dev_salaries, color='#444444', linestyle='--', label='All Devs')
    
ax2.plot(ages, py_salaries, label='Python')
ax2.plot(ages, js_salaries, label='JavaScript')

ax1.legend()
ax1.set_title('Median Salary (USD) by Age')
ax1.set_ylabel('Median Salary (USD)')

ax2.legend()
ax2.set_xlabel('Ages')
ax2.set_ylabel('Median Salary (USD)')

plt.tight_layout()
plt.show()

fig1.savefig('fig1.png') #saving first figure as its own image
fig2.savefig('fig2.png') #saving second figure as its own image
```
